# Adding Artificial Missing values

```python
from random import randint
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/2. edited/letter-recognition - kyan.data")
display(df)
df['lettr'].value_counts().sort_index()
# df.shape
```

### adding Missing values

```python
row_num = df.shape[0]
col_num = df.shape[1]

#  6% missing values
# removed 1 col (target col) since all lettr labels must be present
missing = int(0.06 * (row_num) * (col_num - 1))

# range is from 1 so that lettr col is escaped
for i in range(1, col_num):
    counter = 0
    for _ in range(int(missing/(col_num - 1))):
        df.iloc[randint(0, row_num - 1), i] = np.nan

# convert all columns to int (as they are float after adding missing values)


# --------------- Save the new datase (with missing values) -----------------
# ******* have to save the dataset to be used for later steps (note that since selection of rows is random we can't add missing values multiple times)
df.to_csv('data/2. edited/letter-recognition - kyan - With_NaN.data', index=0)
# # float_format='%.f' -to-> make the dataset col(s) to be int64
# df.to_csv('letter-recognition - kyan - With_NaN.data', index=0, float_format='%.f')

df
```

### check #NaN after operation 

```python
# df.info()

# count the number of Missing values (NaN) per col
display(pd.DataFrame(df.isnull().sum()).transpose())

display('number of rows with missing values:', (df.isnull().any(axis=1)).sum())

# count the number of Missing values (NaN) for all cols
df.isnull().sum().sum()


# plot the Missing values
# to show we did good in spreading the missing values in all features with balance

# set the plot size
fig, ax = plt.subplots(figsize=(10, 15))

sns.heatmap(df.isnull(), cbar=False, ax=ax)
```
